## Plotting with ```matplotlib```

```matplotlib``` is a python library. It contains the ```pyplot``` module, which is basically a collection of functions such as ```plot```, ```title```, ```show()``` etc. ```pyplot``` is one of the most commonly used module for creating a variety of plots such as line plots, bar plots, histograms etc. 

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.linspace(5, 100, 25)
y = np.linspace(10, 1000, 25)

plt.plot(x, y)  #create the plot
plt.xlabel('Profit')
plt.ylabel('Sales')
plt.title('Sales vs Profit for December 2023')
plt.show()   #display the plot
````

````python
x = np.linspace(5, 100, 25)
y = np.linspace(10, 1000, 25)

plt.plot(x, y, 'k+', label = 'x vs y')  #create the plot
plt.xlabel('Profit')
plt.ylabel('Sales')
plt.title('Sales vs Profit for December 2023')
plt.legend()
plt.show()   #display the plot
````

````python
x = np.linspace(5, 100, 25)
y = np.linspace(10, 1000, 25)

plt.plot(x, y, label = 'x vs y')  #create the plot
plt.plot(x, y*2, label = 'x vs y*2')
plt.xlabel('Profit')
plt.ylabel('Sales')
plt.title('Sales vs Profit for December 2023')
plt.legend(loc = 'lower right')
plt.show()   #display the plot
````

## Figures and Subplots

You often need to create multiple plots in the same figure.


```matplotlib``` has the concept of **figures and subplots** using which you can create *multiple subplots inside the same figure*.   

To create multiple plots in the same figure, you can use the method ```plt.subplot(nrows, ncols, nsubplot)```.

````python
merged = pd.read_csv('merged_data.csv')

plt.figure(figsize = (5,12))

plt.subplot(2,1,1)
plt.pie(merged['Product_Category'].value_counts(), labels = merged['Product_Category'].unique(),
       autopct = '%.2f%%', colors = ['Cyan', 'Pink', 'Brown'], explode = [0, 0.3, 0.15])

plt.subplot(2,1,2)
plt.bar(merged['Product_Category'].unique(), height = merged['Product_Category'].value_counts(),
       width = 0.5, color = ['Green', 'Pink', 'Magenta'])
plt.show()
````

Let's see another example - say you want to create 4 subplots in two rows and two columns.

````python
# Example: Create a figure having 4 subplots
x = np.linspace(1, 10, 100)

plt.figure(figsize=(8, 8))
# Optional command, since matplotlib creates a figure by default anyway
#plt.figure(1)

# subplot 1
plt.subplot(2,2,1)
plt.title("Linear")
plt.plot(x, x)

# subplot 2 
plt.subplot(2,2,2)
plt.title("Cubic")
plt.plot(x, x**3)

#plt.figure(figsize=(3, 3))

# subplot 3
#plt.figure(2)

plt.subplot(2,2,3)
plt.title("Log")
plt.plot(x, np.log(x))

# subplot 4 
plt.subplot(2,2,4)
plt.title("Square")
plt.plot(x, x**2)

plt.show()
````

You can see the list of colors and shapes here: https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot

## Types of Plots

**1. Line Plot**

Line plot can be used when we are trying to get the trends over a period of time, eg. Unemployment in India over the last 10 years.

````python
x = np.linspace(0, 10, 20)
y = np.linspace(10, 0, 20)

plt.plot(x,y)

# show the plot
plt.show()
````

**2. Scatter Plot (2 cols which are both numerical(preferrably continous))**

It is effective when we compare one variable across multiple subjects. Ex -  `How much actual sales happened for some marketing budget.`

````python
data = pd.read_csv('marketing_data.csv')
data.dropna(inplace = True)
data
````

````python
x = data['Marketing Budget (X) (In lakhs)']
y = data['Actual Sales(Y) (In crores)']

plt.scatter(x, y, color = 'green', s = 100, marker = '*')
#plt.xlabel()
#plt.ylabel()
#plt.title()
plt.grid()
plt.show()
````

**3. Histogram (1 column that is numerical)**

Histogram is the plot used to show the frequency of continous or discrete variable.

To show distribution of the data for a variable, we can use histogram.

````python
merged.head()
````
