## Merging the Data

```python
market = pd.read_csv('market_fact.csv')
customer = pd.read_csv('cust_dimen.csv')
product = pd.read_csv('prod_dimen.csv')
```

```python
market.head()
```

```python
customer.head()
```

```python
product.head()
```

```python
#pd.merge(market, customer, left_on = 'Cust_id', right_on = 'Id', how = 'inner')
```

```python
new_df = pd.merge(market, customer, on = 'Cust_id', how = 'inner')
new_df.head()
```

```python
final_df = pd.merge(new_df, product, on ='Prod_id', how = 'inner')
final_df.head()
```

```python
final_df.to_csv('merged_data.csv', index = False)
```

## Concatenating the Data

```python
df1 = pd.DataFrame({'Cust_id': [101, 102, 103, 104, 105], 
                   'Prod_id': ['P1', 'P2', 'P3', 'P4', 'P5'],
                   'Profit': [345, 321, 987, 221, 112]})
```

```python
df2 = pd.DataFrame({'Cust_id': [106, 107, 108, 109, 110], 
                   'Prod_id': ['P6', 'P7', 'P8', 'P9', 'P10'],
                   'Profit': [345, 321, 987, 221, 112]})
```

```python
df1
```

## Grouping and Summarizing the Data

```python
data = pd.read_csv('merged_data.csv')
```

```python
data.head()
```

```python
data['Profit'].sum()
```

```python
data['Product_Category'].unique()
```

```python
data.shape
```

```python
data['Product_Category'].value_counts()
```

```python
grp_by_cat = data.groupby('Product_Category')
```

```python
grp_by_cat.groups
```

```python
grp_by_cat['Profit'].sum()  #groupby object
```

```python
pd.DataFrame(grp_by_cat['Profit'].sum()).sort_values('Profit', ascending = False)
```

```python
pd.DataFrame(grp_by_cat['Profit'].sum()).plot.bar()
```

```python
grp_by_cat['Profit'].aggregate(['mean', 'sum', 'min', 'max'])
```

```python
grp_by_cat[['Profit', 'Sales']].sum()
```

```python
grp_by_cat_subcat = data.groupby(['Product_Category', 'Product_Sub_Category'])
```

```python
grp_by_cat_subcat['Profit'].sum()
```
